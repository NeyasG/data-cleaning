```python
import pandas as pd 
#import dataset as a df
df = pd.read_csv(r"C:\Users\henri\OneDrive\Documents\Data Jobs 2023\IO\CV\Projects For Portfolio\Nashville Housing Data for Data Cleaning.csv")

#split owner address column 
df[["owner_street_address","owner_town","owner_state_code"]] = df["OwnerAddress"].str.split(',',2,expand = True)
df[["property_street_address","property_town"]] = df["PropertyAddress"].str.split(',',2,expand = True)

#Split OwnerName into it's three compartments if theres no second owner signalled by lack of '&' fill owner2name column with 'None'
df[['owner_second_name', 'owner1_first_names', 'owner2_first_name']] = df['OwnerName'].str.split(',|&', 2, expand=True)

#Convert YearBuilt column to Int, by replacing NaN values to 0 and then rounding
df['YearBuilt'] = df['YearBuilt'].fillna(0).astype(int)
df['YearBuilt'] = df['YearBuilt'].round().astype(int)
#drop all columns no longer needed 
columns_to_drop = ["OwnerName","HalfBath","FullBath","PropertyAddress"]
df = df.drop(columns = columns_to_drop)

df
```
